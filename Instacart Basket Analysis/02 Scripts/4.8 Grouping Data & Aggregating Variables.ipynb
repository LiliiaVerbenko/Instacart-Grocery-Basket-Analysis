#### Step 7. Determine frequent versus non-frequent customers. Create an order frequency flag that marks the regularity of a userâ€™s ordering behavior according to the median in the â€œdays_since_prior_orderâ€ column. The criteria for the flag should be as follows:
- If the median of â€œdays_since_prior_orderâ€ is higher than 20, then the customer should be labeled a â€œNon-frequent customer.â€
- If the median is higher than 10 and lower than or equal to 20, then the customer should be labeled a â€œRegular customer.â€
- If the median is lower than or equal to 10, then the customer should be labeled a â€œFrequent customer.â€

```python
def frequency_label(row):
    median_value = row['days_since_prior_order']
    
    if median_value > 20:
        return 'Non-frequent customer'
    elif 10 < median_value <= 20:
        return 'Regular customer'
    elif median_value <= 10:
        return 'Frequent customer'
    else:
        return 'NaN'

```

```python
ords_prods_merged['order_frequency_flag'] = ords_prods_merged.apply(frequency_label, axis=1)
```
