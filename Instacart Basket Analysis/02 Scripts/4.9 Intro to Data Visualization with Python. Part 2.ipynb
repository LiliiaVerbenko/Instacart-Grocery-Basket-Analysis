### Creating histograms

Observations:
- The majority of all age groups earn less than 200K
- Across age groups, most earn under 200K. 
- Ages 40+ notices haviing earnings increased  up to 300K, whith significant group earning up to 600K.  
- The income of ages below 40 generally doesn't exceed 400K.
